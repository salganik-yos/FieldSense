```python
# ==========================================
# Cell 1: Environment Setup and Reproducibility
# ==========================================
import os
import torch
import numpy as np
import pynncml as pnc

# Fix random seeds for complete reproducibility across all models
# This ensures that every time we run this split, we get the EXACT same data
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

print("PyNNCML Version:", getattr(pnc, "__version__", "unknown"))
print(f"Random seed set to {RANDOM_SEED} for reproducible data splitting.")
```

```python
# ==========================================
# Cell 2: Load the Entire OpenMRG (Sweden) Dataset
# ==========================================
# Define the full time slice for the project. 
# OpenMRG contains data roughly from June to August 2015.
time_slice = slice("2015-06-01", "2015-08-31")

print(f"Loading full OpenMRG dataset for time slice: {time_slice}...")
full_dataset = pnc.datasets.loader_open_mrg_dataset(time_slice=time_slice)

total_samples = len(full_dataset)
print(f"Dataset loaded successfully.")
print(f"Total time windows (samples) available: {total_samples}")
```
